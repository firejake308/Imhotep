```python
def get_dx_for_sx(sx):
    make_filter = lambda sx, d: ((df_all['RFV1'] == sx) | (df_all['RFV2'] == sx) | (df_all['RFV3'] == sx)) & (~df_all['DIAG'+str(d)].isin(symptomatic_dx))
    x = df_all[make_filter(sx, 1)]['DIAG1'].value_counts()
    y = df_all[make_filter(sx, 2)]['DIAG2'].value_counts()
    z = df_all[make_filter(sx, 3)]['DIAG3'].value_counts()
    return pd.concat([x,y,z], axis=1).sum(axis=1).sort_values(ascending=False)
```
